**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
dfg = df.groupby(by=['London_Borough', 'Year']).mean()
```

```python
dfg.sample(5)
```

```python
#reset the index of dfg
dfg = dfg.reset_index()
```

```python
dfg.head()
```

```python
#create a price ratio of prices from 1995 to 2021
def create_price_ratio(d):
    y1995 = float(d['Average Price'][d['Year']==1995])
    y2021 = float(d['Average Price'][d['Year']==2021])
    ratio = [y1995/y2021]
    return ratio
```

```python
create_price_ratio(dfg[dfg['London_Borough']=='Camden'])
```

```python
#this tells me that in 1995, Camden prices were 14.3% of what they were in 2021
#if we want to see this ratio for all the boroughs, we need to loop through the unique values of the
#London_Borough column and find the 1995 to 2021 ratio
final_ratios = {}
```

```python
for bor in dfg['London_Borough'].unique():
    borough = dfg[dfg['London_Borough']==bor]
    final_ratios[bor] = create_price_ratio(borough)
print(final_ratios)
```

```python
df_ratios = pd.DataFrame(final_ratios)
```

```python
df_ratios.head()
```

```python
#now we need to transpose and reset the index
df_ratios_T = df_ratios.T
df_ratios = df_ratios_T.reset_index()
df_ratios.head()
```

```python
#rename the columns:
#df_ratios.rename(columns={'index': 'Borough', 0: 'Ratio'}, inplace=True)
#df_ratios.head()
df_ratios.rename(columns={'index':'Borough', 0:'2021'}, inplace=True)
df_ratios.head()
```

```python
df_ratios = df_ratios.drop(0)
```

```python
df_ratios.head()
```

```python
#now we want to see the list in descending order, as that will be the largest price change
top20 = df_ratios.sort_values(by='2021', ascending = True).head(20)
print(top20)
```
